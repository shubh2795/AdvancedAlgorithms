```
#Given Recursive function.
#given a large string S and a dictionary of keywords D
#determine if the larger string S may be made up of concatenating a set of strings in the dictionary

# N IS THE NUMBER OF CHARACTERS IN THE STRING
# Name, argument, S and d are read only. n is the count of characters left in S to process
def breakup(n):
  # important n is the number of characters in the string 
  #base case
  # simple problems simple solutions
  # the empty set is contained within the set of all keywords!
  if n == 0:
     return True
  # how can we break the problem into smaller problems?
  # start taking off suffixes. Which ones? all of them
  for i in range(1, n+1):
    #break into S[n-i:n] # suffix and rest S[0: n+1] #rest
    # how do we compute this solution from the subsolutions?
    # return true if this sufix is in the dictionary, and the rest is a solution
    print(" " * (n-i) + S[n-i:n])
    if S[n-i:n] in D and breakup(n-i):
      return True
  return False
     

```

```
#Part 1 :  DP code for breakup.
def breakupDP(n):
  # initialise the cache with all False values
  cache =[False]*(n+1)
  #Fill in the base cases
  cache[0] = True
  # outer loop that iterates from smallest unsolved case to largest unsolved case
  for j in range(1,n+1):
    #Modify the recursive solution construction step and replace recursive calls with cache
    for i in range(1, n+1):
      print(" " * (j-i) + S[j-i:j])
      if S[j-i:j] in D and cache[j-i]:
        cache[j]= True
  return cache[n]

```

```
# Part 2 : Iterative Traceback Routine

def traceback(n):
  
  if n ==0:
    return []

  sol = cache[n]
  for i in range(1, n+1):
    if cache[i]:
      subset = S[n-i:n]
      return subset + traceback(n-i)
  return []
    
```
